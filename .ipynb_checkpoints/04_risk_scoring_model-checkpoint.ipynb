````python
import duckdb
import pandas as pd
import numpy as np
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
from xgboost import XGBClassifier
import mlflow
import mlflow.xgboost
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries loaded!")
````

````python
import duckdb
import pandas as pd
import numpy as np
import yfinance as yf

con = duckdb.connect('C:/Users/munta/Documents/financial-risk-copilot/data/financial_warehouse.duckdb')

tickers = ['AAPL', 'MSFT', 'JPM', 'BAC', 'GS', 'AMZN', 'TSLA', 'XOM', 'JNJ', 'WMT']

# Re-download market data
print("Re-downloading market data...")
market_data = yf.download(tickers, start='2019-01-01', end='2024-12-31', group_by='ticker', auto_adjust=True)

all_data = []
for ticker in tickers:
    df_t = market_data[ticker].copy()
    df_t['ticker'] = ticker
    df_t['date'] = df_t.index
    df_t = df_t.reset_index(drop=True)
    df_t.columns = [c.lower().replace(' ', '_') for c in df_t.columns]
    all_data.append(df_t)

combined = pd.concat(all_data, ignore_index=True).dropna()

# Save market_prices
con.execute("DROP TABLE IF EXISTS market_prices")
con.execute("CREATE TABLE market_prices AS SELECT * FROM combined")
print(f"✅ market_prices saved: {len(combined):,} rows")

# Recreate risk_features using SQL
con.execute("""
    CREATE OR REPLACE TABLE risk_features AS
    SELECT
        ticker,
        date,
        close,
        volume,
        (close - LAG(close) OVER (PARTITION BY ticker ORDER BY date)) 
            / LAG(close) OVER (PARTITION BY ticker ORDER BY date) AS daily_return,
        STDDEV(close) OVER (
            PARTITION BY ticker 
            ORDER BY date 
            ROWS BETWEEN 29 PRECEDING AND CURRENT ROW
        ) AS volatility_30d,
        AVG(close) OVER (
            PARTITION BY ticker 
            ORDER BY date 
            ROWS BETWEEN 6 PRECEDING AND CURRENT ROW
        ) AS ma_7d,
        AVG(close) OVER (
            PARTITION BY ticker 
            ORDER BY date 
            ROWS BETWEEN 29 PRECEDING AND CURRENT ROW
        ) AS ma_30d
    FROM market_prices
    ORDER BY ticker, date
""")

# Load for modeling
df = con.execute("""
    SELECT * FROM risk_features
    WHERE daily_return IS NOT NULL
    AND volatility_30d IS NOT NULL
    ORDER BY ticker, date
""").df()

con.close()

print(f"✅ risk_features saved: {len(df):,} rows")
print(df.head())
````

````python
# Create binary risk label:
# 1 = HIGH RISK (price dropped more than 2% next day)
# 0 = LOW RISK

df = df.sort_values(['ticker', 'date']).reset_index(drop=True)

# Next day return
df['next_return'] = df.groupby('ticker')['daily_return'].shift(-1)

# Risk label
df['risk_label'] = (df['next_return'] < -0.02).astype(int)

# Additional features
df['return_7d'] = df.groupby('ticker')['daily_return'].transform(
    lambda x: x.rolling(7).mean()
)
df['volatility_7d'] = df.groupby('ticker')['daily_return'].transform(
    lambda x: x.rolling(7).std()
)
df['price_vs_ma7'] = (df['close'] - df['ma_7d']) / df['ma_7d']
df['price_vs_ma30'] = (df['close'] - df['ma_30d']) / df['ma_30d']
df['volume_change'] = df.groupby('ticker')['volume'].pct_change()

# Drop nulls
df = df.dropna()

print(f"Final dataset: {len(df):,} rows")
print(f"Risk label distribution:\n{df['risk_label'].value_counts()}")
````

````python
# Feature columns
features = [
    'daily_return',
    'volatility_30d',
    'volatility_7d', 
    'return_7d',
    'price_vs_ma7',
    'price_vs_ma30',
    'volume_change',
    'close',
    'volume'
]

X = df[features]
y = df['risk_label']

# Time-based split (NEVER random split for time series!)
split_idx = int(len(df) * 0.8)
X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"Training set: {len(X_train):,} rows")
print(f"Test set:     {len(X_test):,} rows")
````

````python
# Set up MLflow
mlflow.set_tracking_uri('C:/Users/munta/Documents/financial-risk-copilot/mlruns')
mlflow.set_experiment("financial-risk-scoring")

with mlflow.start_run(run_name="xgboost_baseline"):
    
    # Model parameters
    params = {
        'n_estimators': 200,
        'max_depth': 6,
        'learning_rate': 0.05,
        'subsample': 0.8,
        'colsample_bytree': 0.8,
        'scale_pos_weight': 3,  # handle class imbalance
        'random_state': 42,
        'eval_metric': 'logloss'
    }
    
    # Train model
    model = XGBClassifier(**params)
    model.fit(
        X_train_scaled, y_train,
        eval_set=[(X_test_scaled, y_test)],
        verbose=False
    )
    
    # Evaluate
    y_pred = model.predict(X_test_scaled)
    y_proba = model.predict_proba(X_test_scaled)[:, 1]
    auc = roc_auc_score(y_test, y_proba)
    
    # Log to MLflow
    mlflow.log_params(params)
    mlflow.log_metric("roc_auc", auc)
    mlflow.xgboost.log_model(model, "xgboost_risk_model")
    
    print(f"ROC-AUC Score: {auc:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, 
          target_names=['Low Risk', 'High Risk']))
````

````python
import plotly.graph_objects as go

# Get feature importances
importance_df = pd.DataFrame({
    'feature': features,
    'importance': model.feature_importances_
}).sort_values('importance', ascending=True)

# Plot
fig = go.Figure(go.Bar(
    x=importance_df['importance'],
    y=importance_df['feature'],
    orientation='h',
    marker_color='steelblue'
))

fig.update_layout(
    title='XGBoost Feature Importance — Financial Risk Model',
    xaxis_title='Importance Score',
    yaxis_title='Feature',
    height=400
)

fig.show()
````

````python
# Score all companies
X_all_scaled = scaler.transform(X[features])
df['risk_score'] = model.predict_proba(X_all_scaled)[:, 1]
df['risk_category'] = pd.cut(
    df['risk_score'],
    bins=[0, 0.3, 0.6, 1.0],
    labels=['Low Risk', 'Medium Risk', 'High Risk']
)

# Latest risk score per company
latest_risk = df.sort_values('date').groupby('ticker').last()[
    ['close', 'risk_score', 'risk_category', 'volatility_30d']
].reset_index()

latest_risk = latest_risk.sort_values('risk_score', ascending=False)
print("Latest Risk Scores by Company:")
print(latest_risk.to_string(index=False))
````
